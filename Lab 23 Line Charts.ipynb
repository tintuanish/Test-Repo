### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
# 1. Group by 'Age' and calculate the median of 'ConvertedCompYearly'
df_comp_age = df.groupby('Age')['ConvertedCompYearly'].median()

# 2. Plot the line chart
plt.figure(figsize=(12, 6))
df_comp_age.plot(kind='line', marker='o', linestyle='-', color='b')

plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
# 1. Convert the experience column to numeric (handle 'Less than 1 year' as 0 and 'More than 50 years' as 51)
df['YearsCodePro_num'] = pd.to_numeric(df['YearsCodePro'].replace({'Less than 1 year': 0, 'More than 50 years': 51}), errors='coerce')

# 2. Group by experience and calculate the median Job Satisfaction
df_exp_sat = df.groupby('YearsCodePro_num')['JobSatPoints_6'].median().sort_index()

# 3. Plot the trend
plt.figure(figsize=(12, 6))
df_exp_sat.plot(kind='line', marker='o', color='purple')

plt.title('Median Job Satisfaction by Years of Professional Experience')
plt.xlabel('Years of Professional Experience')
plt.ylabel('Median Job Satisfaction')
plt.grid(True)
plt.show()
```
