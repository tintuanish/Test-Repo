## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
# 1. Count the number of duplicate rows in the dataset
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# 2. Display the first few duplicate rows to understand their structure
duplicate_rows = df[df.duplicated()]
duplicate_rows.head()
```

### Task 2: Analysis of Duplicate Response Patterns
1. Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.
2. Clarify that these represent multiple respondents with identical answers rather than duplicate records. Analyse which other columns frequently share identical values within these response-pattern groups.
   


```python
## Write your code here
# 1. Identify duplicate response patterns based on selected columns
selected_columns = ['MainBranch', 'Employment', 'RemoteWork']
pattern_duplicates = df.duplicated(subset=selected_columns).sum()
print(f"Number of rows sharing patterns in {selected_columns}: {pattern_duplicates}\n")

# 2. Analyze which other columns frequently share identical values within these groups
# Grouping by the pattern and checking for identical values across other columns
grouped = df.groupby(selected_columns)

print("Columns where respondents within the same pattern frequently share identical values:")
# We look for columns that have exactly 1 unique value within each group on average
for col in df.columns:
    if col not in selected_columns:
        # Calculate the average number of unique values per group for this column
        avg_unique = grouped[col].nunique().mean()
        if avg_unique == 1:
            print(f" - {col} (Always identical within the groups)")
        elif avg_unique < 1.5:
            print(f" - {col} (Highly identical, avg unique values: {avg_unique:.2f})")
```

### Task 3: Visualize Shared Response Patterns
1. Create visualizations to show the distribution of shared response patterns across different categories.
2. Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment.


```python
## Write your code here
!pip install seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Fix: We don't manually add 'Employment' because it's already inside selected_columns
grouped_patterns = df.groupby(['Country'] + selected_columns).size().reset_index(name='Count')

# Grab the top 10 combinations for a clear visualization
top_patterns = grouped_patterns.sort_values(by='Count', ascending=False).head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_patterns, 
    x='Count', 
    y='Country', 
    hue='RemoteWork', 
    palette='viridis'
)

plt.title('Top 10 Shared Response Patterns by Country and Remote Work Status')
plt.xlabel('Number of Respondents')
plt.ylabel('Country')
plt.legend(title='Remote Work', loc='lower right')
plt.tight_layout()
plt.show()
```

### Task 4: Evaluate Duplicate Handling Strategy
1. Analyse the dataset to determine which column(s) define record uniqueness.
2. Assess whether removing rows based on a subset of columns (rather than complete row duplication) is appropriate.
Justify your decision with reference to the structure and purpose of the dataset.


## Write your analysis here
Justification of Duplicate Handling Strategy
Defining Record Uniqueness: The true uniqueness of a record in a survey dataset is determined by its unique identifier column (e.g., Respondent or ResponseId). If two rows share identical respondent IDs, they are genuine data duplicates that should be removed.

Subset-Based Deduplication Assessment: Removing rows based only on a subset of demographic or behavioral columns (such as a shared pattern of MainBranch, Employment, and RemoteWork) is inappropriate and highly risky.

Reasoning: In a large-scale international developer survey, it is entirely expected and statistically guaranteed that thousands of distinct individuals will share identical high-level characteristics (e.g., being a full-time employed remote developer in the United States). Dropping these rows simply because they share a common response pattern would strip away authentic survey data, skew the demographic representation, and completely invalidate any statistical conclusions drawn from the dataset. Therefore, deduplication must only be applied to identical complete rows or duplicate unique identifiers.


```python
## Verify and Document Duplicate Removal Process
# Automatically find and check the unique identifier column
id_column = None
for col in df.columns:
    if 'id' in col.lower() or 'respondent' in col.lower():
        id_column = col
        break

if id_column:
    is_unique = df[id_column].is_unique
    unique_count = df[id_column].nunique()
    total_rows = len(df)
    
    print(f"Found ID Column: '{id_column}'")
    print(f"Total rows in dataset: {total_rows}")
    print(f"Number of unique IDs: {unique_count}")
    print(f"Is the ID column perfectly unique? {is_unique}")
else:
    print("Could not automatically find an ID column. Here are the columns you have:")
    print(df.columns.tolist())
```
