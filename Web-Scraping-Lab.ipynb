## Extract information from the given web site
You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
# Your code here
```

Download the webpage at the url


```python
#your code goes here
# The url is already defined in the cell above, but included here for completeness
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"

# Download the webpage
data = requests.get(url).text
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data, "html.parser")
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
# Initialize an empty list to store the scraped data
languages_data = []

# Find the table in the webpage
table = soup.find("table")

# Loop through the rows of the table (skipping the header row)
for row in table.find_all("tr")[1:]:
    cols = row.find_all("td")
    language_name = cols[1].getText().strip()
    average_salary = cols[3].getText().strip()

    # Append the data as a dictionary to our list
    languages_data.append(
        {"Language": language_name, "Average Annual Salary": average_salary}
    )

# Quick look at the scraped data
print(languages_data[:5])
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
# Create a DataFrame from the scraped list
df = pd.DataFrame(languages_data)

# Save the DataFrame to a CSV file
df.to_csv("popular-languages.csv", index=False)

print("Data successfully saved to popular-languages.csv")
```
