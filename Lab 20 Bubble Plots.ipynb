### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
!pip install seaborn
```

```python
##Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Prepare the data
# drop rows where Age or SOPartFreq are missing
df_task1_1 = df.dropna(subset=['Age', 'SOPartFreq', 'JobSat'])
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_task1_1, 
    x='Age', 
    y='SOPartFreq', 
    size='JobSat', 
    sizes=(20, 200), 
    alpha=0.5,
    legend=False
)
plt.title('Age vs. Frequency of Participation (Size = Job Satisfaction)')
plt.xticks(rotation=45)
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
# Prepare the data
# Dropping rows where compensation or job satisfaction are missing
df_task2_2 = df.dropna(subset=['ConvertedCompYearly', 'JobSat'])

# Create the bubble plot
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_task2_2, 
    x='ConvertedCompYearly', 
    y='JobSat', 
    size='Age',          # Using Age for bubble size as requested
    sizes=(20, 200), 
    alpha=0.5,
    legend=False
)
plt.title('Compensation vs. Job Satisfaction (Size = Age)')
plt.xticks(rotation=45)
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
# Aggregate data for top 15 languages
lang_counts = df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack().value_counts()
top_15 = lang_counts.head(15).reset_index()
top_15.columns = ['Language', 'Count']

# Create the Bubble Plot
# Using a scatterplot with sizes based on the Count (frequency)
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=top_15, 
    x='Count', 
    y='Language', 
    size='Count', 
    sizes=(100, 1000), 
    hue='Language',
    legend=False
)

plt.title('Top 15 Most Popular Programming Languages')
plt.xlabel('Number of Respondents')
plt.ylabel('Programming Language')
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
# 1. Filter and prepare data
df_2_2 = df.dropna(subset=['DatabaseWantToWorkWith', 'JobSat'])

# 2. To avoid clutter, let's look at the top 10 most common databases
top_dbs = df_2_2['DatabaseWantToWorkWith'].value_counts().nlargest(10).index
df_2_2_filtered = df_2_2[df_2_2['DatabaseWantToWorkWith'].isin(top_dbs)]

# 3. Create the Bubble Plot
# We group by Database and Job Satisfaction to get counts for bubble size
df_bubble = df_2_2_filtered.groupby(['DatabaseWantToWorkWith', 'JobSat']).size().reset_index(name='Count')

plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=df_bubble, 
    x='JobSat', 
    y='DatabaseWantToWorkWith', 
    size='Count', 
    sizes=(100, 1000), 
    hue='DatabaseWantToWorkWith',
    alpha=0.6,
    legend=False
)

plt.title('Preferred Databases vs. Job Satisfaction (Top 10)')
plt.xlabel('Job Satisfaction')
plt.ylabel('Database')
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
# 1. Prepare data
# We split the DevType string to handle multiple roles and drop missing values
df_dev = df.dropna(subset=['ConvertedCompYearly', 'DevType', 'JobSat'])
df_dev_split = df_dev.assign(DevType=df_dev['DevType'].str.split(';')).explode('DevType')

# 2. To keep it clean, let's group by DevType and take the average of Comp and JobSat
df_bubble = df_dev_split.groupby('DevType').agg({
    'ConvertedCompYearly': 'mean',
    'JobSat': 'mean'
}).reset_index()

# 3. Create the Bubble Plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=df_bubble, 
    x='ConvertedCompYearly', 
    y='DevType', 
    size='JobSat', 
    sizes=(100, 1000), 
    hue='DevType',
    alpha=0.6,
    legend=False
)

plt.title('Average Compensation vs. Job Satisfaction by Role')
plt.xlabel('Average Converted Compensation (Yearly)')
plt.ylabel('Developer Role')
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
# 1. Prepare data
# Split the tools and drop missing values
df_tools = df.dropna(subset=['NEWCollabToolsHaveWorkedWith', 'Age'])
df_tools_split = df_tools.assign(
    Tool=df_tools['NEWCollabToolsHaveWorkedWith'].str.split(';')
).explode('Tool')

# 2. Group by Age and Tool to get counts for bubble size
df_bubble = df_tools_split.groupby(['Age', 'Tool']).size().reset_index(name='Count')

# 3. Create the Bubble Plot
plt.figure(figsize=(14, 10))
sns.scatterplot(
    data=df_bubble, 
    x='Age', 
    y='Tool', 
    size='Count', 
    sizes=(50, 800), 
    hue='Tool',
    alpha=0.6,
    legend=False
)

plt.title('Collaboration Tools Used by Age Group')
plt.xlabel('Age')
plt.ylabel('Collaboration Tool')
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
# 1. Filter and prepare data
# We drop rows where Country or LanguageAdmired is missing
df_4_2 = df.dropna(subset=['Country', 'LanguageAdmired'])

# 2. To keep the bubble plot readable, let's focus on the Top 10 Countries 
# and Top 10 most admired languages
top_countries = df_4_2['Country'].value_counts().nlargest(10).index
top_languages = df_4_2['LanguageAdmired'].value_counts().nlargest(10).index

df_bubble = df_4_2[
    (df_4_2['Country'].isin(top_countries)) & 
    (df_4_2['LanguageAdmired'].isin(top_languages))
].groupby(['Country', 'LanguageAdmired']).size().reset_index(name='Count')

# 3. Create the Bubble Plot
plt.figure(figsize=(14, 10)) # Increased size to stop warning
sns.scatterplot(
    data=df_bubble, 
    x='Country', 
    y='LanguageAdmired', 
    size='Count', 
    sizes=(100, 1500), 
    hue='LanguageAdmired',
    alpha=0.6,
    legend=False
)

plt.title('Admired Technologies Across Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Admired Technology')
plt.xticks(rotation=45, ha='right') # Rotated labels to prevent overlap
plt.tight_layout()
plt.show()
```
