### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

```python
# Write your code here
# Generate basic statistics for all numerical columns
print("Basic statistics for numerical columns:")
df.describe()
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
# 1. Check for the number of duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows before removal: {duplicate_count}")

# 2. Remove duplicate rows if any exist
if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print(f"Successfully removed duplicates. New duplicate count: {df.duplicated().sum()}")
else:
    print("No duplicate rows found to remove.")
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
# 1. Strip leading and trailing whitespaces from the Country column
df['Country'] = df['Country'].astype(str).str.strip()

# 2. Display the top 10 unique country entries to check their format
print("Top 10 unique countries after cleaning whitespaces:")
print(df['Country'].value_counts().head(10))
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
# 1. Apply One-Hot Encoding to the Employment column
df_encoded = pd.get_dummies(df, columns=['Employment'], prefix='Emp', dtype=int)

# 2. Display the new columns to verify the encoding
emp_columns = [col for col in df_encoded.columns if col.startswith('Emp_')]
print("Newly created One-Hot Encoded columns:")
print(emp_columns)

# 3. Preview the first few rows of these new columns
print("\nPreview of encoded columns:")
print(df_encoded[emp_columns].head())

# Update our primary dataframe reference if you want to keep the changes moving forward
df = df_encoded
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
# 1. Calculate the number of missing values per column
missing_data = df.isnull().sum()

# 2. Filter out columns that have 0 missing values and sort them in descending order
missing_data_filtered = missing_data[missing_data > 0].sort_values(ascending=False)

# 3. Display the top 10 columns with the highest number of missing values
print("Top 10 columns with the highest number of missing values:")
print(missing_data_filtered.head(10))
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
# 1. Calculate the median of the ConvertedCompYearly column
median_comp = df['ConvertedCompYearly'].median()
print(f"Calculated median salary: {median_comp}")

# 2. Impute missing values with the median
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_comp)

# 3. Verify that there are no more missing values left in this column
print(f"Missing values in 'ConvertedCompYearly' after imputation: {df['ConvertedCompYearly'].isnull().sum()}")
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# 1. Find the most frequent value (mode) in the RemoteWork column
remote_work_mode = df['RemoteWork'].mode()[0]
print(f"Most frequent RemoteWork status: {remote_work_mode}")

# 2. Impute missing values with the mode
df['RemoteWork'] = df['RemoteWork'].fillna(remote_work_mode)

# 3. Verify that there are no more missing values left in this column
print(f"Missing values in 'RemoteWork' after imputation: {df['RemoteWork'].isnull().sum()}")
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
# 1. Calculate minimum and maximum values of the column
min_val = df['ConvertedCompYearly'].min()
max_val = df['ConvertedCompYearly'].max()

# 2. Apply Min-Max Scaling formula
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - min_val) / (max_val - min_val)

# 3. Verify the scaling by printing the new min and max (should be 0.0 and 1.0)
print("Min-Max Scaling Verification:")
print(f"Minimum value: {df['ConvertedCompYearly_MinMax'].min()}")
print(f"Maximum value: {df['ConvertedCompYearly_MinMax'].max()}")
```
