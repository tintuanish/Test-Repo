### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# 1. Split the semicolon-separated values and count them
# The explode function turns each list element into its own row
data = df['DatabaseWantToWorkWith'].str.split(';').explode()
top_5_databases = data.value_counts().head(5)

# 2. Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
# Create the pie chart for DevType
data = df['DevType'].str.split(';').explode()
top_5_devtype = data.value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(top_5_devtype, labels=top_5_devtype.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Developer Types')
plt.show()
```
