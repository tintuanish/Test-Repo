#### Step 3: Handling Missing Data


Identify and manage missing values in critical columns such as `Employment`, `JobSat`, and `RemoteWork`. Implement a strategy to fill or drop these values, depending on the significance of the missing data.


```python
## Write your code here
# 1. Check the number of missing values in the critical columns
print("Missing values before cleaning:")
print(df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum())

# 2. Drop rows where any of these critical columns have missing values
df.dropna(subset=['Employment', 'JobSat', 'RemoteWork'], inplace=True)

# 3. Verify that the missing values have been removed
print("\nMissing values after cleaning:")
print(df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum())
```

#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
## Write your code here
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Clean and convert YearsCodePro to numeric format
df['YearsCodePro'] = pd.to_numeric(
    df['YearsCodePro'].astype(str).str.replace('Less than 1 year', '0').str.replace('More than 50 years', '51'), 
    errors='coerce'
)

# 2. Directly convert JobSat to a numeric score (handling pre-existing numeric strings)
df['JobSat_Numeric'] = pd.to_numeric(df['JobSat'], errors='coerce')

# 3. Create the experience ranges using pd.cut
bins = [-1, 5, 10, 20, float('inf')]
labels = ['0-5 years', '5-10 years', '10-20 years', '>20 years']
df['Experience_Range'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels)

# 4. Calculate the median JobSat score for each range
median_satisfaction = df.dropna(subset=['JobSat_Numeric', 'Experience_Range']).groupby('Experience_Range', observed=False)['JobSat_Numeric'].median().reset_index()
print("Median Job Satisfaction Score by Experience Range:")
print(median_satisfaction)

# 5. Visualize the relationship using a bar plot (with hue assigned to avoid future warnings)
plt.figure(figsize=(8, 5))
sns.barplot(x='Experience_Range', y='JobSat_Numeric', data=median_satisfaction, hue='Experience_Range', palette='Blues_d', legend=False)
plt.title('Median Job Satisfaction by Experience Range')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction Score')
plt.ylim(1, df['JobSat_Numeric'].max() if not df['JobSat_Numeric'].isna().all() else 5)
plt.show()
```

#### Step 5: Visualize Job Satisfaction


Use a count plot to show the distribution of `JobSat` values. This provides insights into the overall satisfaction levels of respondents.


```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Sort the categories or numeric values to ensure a clean chronological x-axis
df_clean_sat = df.dropna(subset=['JobSat_Numeric']).sort_values('JobSat_Numeric')

# Create a count plot to show the overall distribution of JobSat
plt.figure(figsize=(10, 5))
sns.countplot(x='JobSat_Numeric', data=df_clean_sat, hue='JobSat_Numeric', palette='viridis', legend=False)

plt.title('Overall Distribution of Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Count of Respondents')
plt.show()
```

#### Step 6: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
## Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Clean up any missing values in RemoteWork or Employment for this analysis
df_remote = df.dropna(subset=['RemoteWork', 'Employment']).copy()

# 2. Simplify the Employment category to eliminate overcrowding and axis errors
def simplify_employment(emp_string):
    if 'full-time' in emp_string.lower():
        return 'Full-time Employee'
    elif 'independent contractor' in emp_string.lower() or 'freelancer' in emp_string.lower():
        return 'Independent/Freelancer'
    elif 'part-time' in emp_string.lower():
        return 'Part-time Employee'
    elif 'student' in emp_string.lower():
        return 'Student'
    else:
        return 'Other'

df_remote['Employment_Clean'] = df_remote['Employment'].apply(simplify_employment)

# 3. Plot the overall distribution of remote work preferences (with hue set)
plt.figure(figsize=(8, 5))
sns.countplot(data=df_remote, x='RemoteWork', hue='RemoteWork', palette='Set2', legend=False)
plt.title('Overall Distribution of Remote Work Preferences')
plt.xlabel('Remote Work Status')
plt.ylabel('Count of Respondents')
plt.show()

# 4. Cross-tabulate the simplified, clean categories
print("Simplified Cross-tabulation of Remote Work by Employment Type:")
crosstab_res = pd.crosstab(df_remote['Employment_Clean'], df_remote['RemoteWork'])
print(crosstab_res)

# 5. Visualize the simplified cross-tabulation using a grouped bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=df_remote, x='Employment_Clean', hue='RemoteWork', palette='viridis')
plt.title('Remote Work Preferences broken down by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Count of Respondents')
plt.legend(title='Remote Work Preference')
plt.tight_layout()  # This will now execute perfectly without warnings!
plt.show()
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
## Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Drop rows missing Language or Country data
df_lang = df.dropna(subset=['LanguageHaveWorkedWith', 'Country']).copy()

# 2. Find the top 3 most represented countries in the dataset to focus our analysis
top_countries = df_lang['Country'].value_counts().nlargest(3).index.tolist()
df_top_regions = df_lang[df_lang['Country'].isin(top_countries)].copy()

# 3. Split the semi-colon separated languages into individual rows
df_top_regions['Language'] = df_top_regions['LanguageHaveWorkedWith'].str.split(';')
df_exploded = df_top_regions.explode('Language')

# 4. Find the top 5 languages for each of these top regions
top_languages_by_region = (
    df_exploded.groupby(['Country', 'Language'], observed=False)
    .size()
    .reset_index(name='Count')
    .sort_values(['Country', 'Count'], ascending=[True, False])
    .groupby('Country')
    .head(5)
)

print("Top 5 Programming Languages by Top Regions:")
print(top_languages_by_region)

# 5. Visualize using a grouped bar plot (explicitly setting hue to avoid future warnings)
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_languages_by_region, 
    x='Country', 
    y='Count', 
    hue='Language', 
    palette='tab10'
)

plt.title('Top 5 Programming Languages Across Major Regions')
plt.xlabel('Region / Country')
plt.ylabel('Number of Responses')
plt.legend(title='Programming Language', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### Step 8: Correlation Between Experience and Satisfaction


Examine how years of experience (`YearsCodePro`) correlate with job satisfaction (`JobSatPoints_1`). Use a scatter plot to visualize this relationship.


```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Clean up missing values for our two targeted columns
# We'll check if JobSatPoints_1 is in the dataframe, otherwise use our engineered JobSat_Numeric
sat_col = 'JobSatPoints_1' if 'JobSatPoints_1' in df.columns else 'JobSat_Numeric'
df_corr = df.dropna(subset=['YearsCodePro', sat_col])

# 2. Visualize the relationship using a scatter plot with a trend line (regplot)
plt.figure(figsize=(10, 6))
sns.regplot(
    data=df_corr, 
    x='YearsCodePro', 
    y=sat_col, 
    scatter_kws={'alpha':0.3, 'color': '#1f77b4'}, 
    line_kws={'color': 'red', 'linewidth': 2}
)

plt.title('Correlation Between Professional Coding Experience and Job Satisfaction')
plt.xlabel('Years of Professional Coding Experience (YearsCodePro)')
plt.ylabel('Job Satisfaction Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 3. Print out the mathematical correlation coefficient
correlation = df_corr['YearsCodePro'].corr(df_corr[sat_col])
print(f"Pearson Correlation Coefficient: {correlation:.4f}")
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
## Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Clean up missing values for our target columns
df_edu = df.dropna(subset=['EdLevel', 'Employment']).copy()

# 2. Simplify EdLevel strings to make the plot readable
def simplify_edu(edu_string):
    edu_lower = str(edu_string).lower()
    if 'bachelor' in edu_lower:
        return "Bachelor's"
    elif 'master' in edu_lower:
        return "Master's"
    elif 'professional degree' in edu_lower or 'phd' in edu_lower or 'doctor' in edu_lower:
        return 'Doctorate/Professional'
    elif 'associate' in edu_lower or 'secondary' in edu_lower or 'high school' in edu_lower:
        return 'High School/Associate'
    else:
        return 'Other'

df_edu['EdLevel_Clean'] = df_edu['EdLevel'].apply(simplify_edu)

# 3. Simplify Employment strings (using our logic from Step 6)
def simplify_employment(emp_string):
    emp_lower = str(emp_string).lower()
    if 'full-time' in emp_lower:
        return 'Full-time'
    elif 'independent contractor' in emp_lower or 'freelancer' in emp_lower:
        return 'Freelancer/Contractor'
    elif 'part-time' in emp_lower:
        return 'Part-time'
    else:
        return 'Other'

df_edu['Employment_Clean'] = df_edu['Employment'].apply(simplify_employment)

# 4. Generate the Cross-tabulation table
print("Cross-tabulation of Education Level vs Employment Type:")
edu_crosstab = pd.crosstab(df_edu['EdLevel_Clean'], df_edu['Employment_Clean'], normalize='index') * 100
print(edu_crosstab.round(2))

# 5. Visualize with a structured, warning-free stacked bar plot
edu_crosstab.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Employment Type Breakdown by Educational Background')
plt.xlabel('Education Level')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=15)
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
