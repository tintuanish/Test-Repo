### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
##Write your code here
# 1. Prepare the data: Group by Age and calculate the median
df_task1_1 = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# 2. Plot the stacked bar chart
df_task1_1.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Median Job Satisfaction Points (6 & 7) by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Satisfaction Points')
plt.tight_layout()
plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
##Write your code here
# 1. Prepare the data
# We group by Age and take the median of Compensation and Job Satisfaction
df_task2 = df.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

# 2. Plotting
# Note: Since Comp and Satisfaction have very different scales, 
# you might need to normalize them or plot them on secondary axes.
# Here is the standard stacked approach:
df_task2.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Median Compensation & Job Satisfaction by Age')
plt.xlabel('Age Group')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
##Write your code here
# 1. Prepare the data
# Grouping by Age and calculating the median for both metrics
df_task2_1 = df.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

# 2. Plotting with a secondary axis for better visibility
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Compensation on the primary y-axis
color1 = 'tab:blue'
ax1.set_xlabel('Age Group')
ax1.set_ylabel('Median Compensation', color=color1)
df_task2_1['ConvertedCompYearly'].plot(kind='bar', color=color1, ax=ax1, alpha=0.6, position=1, width=0.4)
ax1.tick_params(axis='y', labelcolor=color1)

# Create a second y-axis for Job Satisfaction
ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Median Job Satisfaction (Points 6)', color=color2)
df_task2_1['JobSatPoints_6'].plot(kind='bar', color=color2, ax=ax2, alpha=0.6, position=0, width=0.4)
ax2.tick_params(axis='y', labelcolor=color2)

plt.title('Median Compensation vs. Job Satisfaction by Age')
plt.tight_layout()
plt.show()
```

##### Stacked Chart of Median Compensation and Job Satisfaction Across Age Group


Compare the median compensation and job satisfaction metrics across different age groups. This helps visualize how compensation and satisfaction levels differ by age.


```python
##Write your code here
# 1. Prepare the data
df_emp_sat_pivot = df.groupby(['Employment', 'JobSatPoints_6']).size().unstack()

# 2. Plotting with extra margin (subplots_adjust)
fig, ax = plt.subplots(figsize=(14, 7))
df_emp_sat_pivot.plot(kind='bar', stacked=True, ax=ax)

plt.title('Employment Type by Job Satisfaction Level')
plt.xlabel('Employment Status')
plt.ylabel('Count of Respondents')
plt.legend(title='Satisfaction Points', bbox_to_anchor=(1.02, 1), loc='upper left', fontsize='small')

# Manually adjust the margins to give the layout room to breathe
plt.subplots_adjust(right=0.85) 

plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
##Write your code here
# 1. Prepare the data
# Filter for top databases to keep the chart readable
top_databases = df['DatabaseAdmired'].value_counts().nlargest(8).index
df_db = df[df['DatabaseAdmired'].isin(top_databases)]

# Group by Age and Database, then count
df_db_pivot = df_db.groupby(['Age', 'DatabaseAdmired']).size().unstack()

# 2. Plotting
fig, ax = plt.subplots(figsize=(14, 7))
df_db_pivot.plot(kind='bar', stacked=True, ax=ax)

plt.title('Preferred Databases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count of Respondents')
plt.legend(title='Database', bbox_to_anchor=(1.02, 1), loc='upper left')

plt.subplots_adjust(right=0.85)
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
##Write your code here
# 1. Prepare the data
# Filter for top 8 programming languages
top_langs = df['LanguageAdmired'].value_counts().nlargest(8).index
df_langs = df[df['LanguageAdmired'].isin(top_langs)]

# Group by Age and Language, then count
df_langs_pivot = df_langs.groupby(['Age', 'LanguageAdmired']).size().unstack()

# 2. Plotting
fig, ax = plt.subplots(figsize=(14, 7))
df_langs_pivot.plot(kind='bar', stacked=True, ax=ax)

plt.title('Preferred Programming Languages by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count of Respondents')
plt.legend(title='Language', bbox_to_anchor=(1.02, 1), loc='upper left')

# Adjust layout to fit the legend
plt.subplots_adjust(right=0.85)
plt.show()
```
